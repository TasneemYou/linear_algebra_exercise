```python
import matplotlib.pyplot as plt
import numpy as np

from PIL import Image
```

```python
img = Image.open('flower_square.jpg')
imggray = img.convert('LA') #convert it to black-and-white
plt.figure(figsize=(9,6))
plt.imshow(imggray)
```

```python
imggray.size
```

```python
imgmat = np.array(list(imggray.getdata(band=0)),float)  #imggray.getdata() Returns the contents of this image as a sequence object
                                                        #containing pixel values of band 0 (i.e. R from RGB).
   #The sequence object is flattened, so the values for line one follow directly after the values of line zero, and so on.
type(imgmat)
```

```python
imgmat.shape
```

```python
imgmat.shape = (imggray.size[1],imggray.size[0])
```

```python
imgmat.shape
```

```python
imgmat = np.matrix(imgmat)
```

```python
imgmat.shape
```

```python
plt.figure(figsize=(9,6))
plt.imshow(imgmat, cmap = 'gray')
```

```python
U, sigma, V = np.linalg.svd(imgmat)  # create SVD of image
```

```python
reconstrimg = np.matrix(U[:,:1]) * np.diag(sigma[:1]) * np.matrix(V[:1,:]) # approximation (reconstruction) of image using
                                                                #only 1st column of U and 1st row of V
plt.imshow(reconstrimg, cmap = 'gray')
```

```python
for i in range(2,4):
    reconstrimg = np.matrix(U[:,:i])*np.diag(sigma[:i])*np.matrix(V[:i,:])
    plt.imshow(reconstrimg, cmap = 'gray')
    title = 'n = %s' %i
    plt.title(title)
    plt.show()
```

```python
for i in range(5,51,4):
    reconstrimg = np.matrix(U[:,:i])*np.diag(sigma[:i])*np.matrix(V[:i,:])
    plt.imshow(reconstrimg, cmap = 'gray')
    title = 'n = %s' %i
    plt.title(title)
    plt.show()
```
